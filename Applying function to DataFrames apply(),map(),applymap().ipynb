```python
df = pd.DataFrame({ 'A': [1,2,3,4], 
                   'B': [10,20,30,40],
                   'C': [20,40,60,80]
                  }, 
                  index=['Row 1', 'Row 2', 'Row 3', 'Row 4'])
df
```

### Apply() function

```python
### Here axis = 1 means 'row' and axis=0 means 'column' 
def custom(y):
    return y.sum()
df['D'] = df.apply(custom, axis=1)
```

```python
df
```

```python
df.loc['Row 5'] = df.apply(custom, axis=0)
df
```

```python
df['E'] = df.apply(lambda x: x['C'] * x['D'], axis=1)
```

```python
df
```

```python
### Use this for applying on series.

def multiply_by_2(val):
    return val * 2
df['C'] = df['C'].apply(multiply_by_2)
df
```

### map() function 
Note. Very useful for converting categorical columns to numeric .
Used to subtitute each value with another value.

```python
import numpy as np
ser = pd.DataFrame( {'x':['cat', 'dog', 'pigon', 'rabbit'],
                    'y':['bird', 'dog', 'snake', 'eagle']})
ser
```

```python
### map() accepts a dictionary or series or function.
###  Values that are not found in the dict are converted to NaN,
ser['x'].map({'cat':1, 'dog':2})
```

```python
varlist = ['x','y']
def yyy(c):
    return c.map({'cat':33, 'dog':44, 'pigon':55 , 'rabit':66 , 'bird':100 , 'snake':500})
 
ser[varlist] = ser[varlist].apply(yyy) 
```

```python
ser
```

```python
ser['x'].map('Im a {}'.format, na_action='ignore')
```
